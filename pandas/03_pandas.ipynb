```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', np.nan],
    'Age': [25, np.nan, 30, 22, 28],
    'City': ['New York', 'Los Angeles', np.nan, 'Chicago', 'Houston'],
    'Score': [85, 90, np.nan, 88, 92]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()

df.dropna(axis=1 ,inplace=True)
df
```

```python
# You must specify a value or method to fill NaNs, e.g., fill with 0
df.fillna(value=0, axis=1, inplace=True)
df
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Age"].fillna(df["Age"].mean(), inplace=True)
df
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df
```

```python
df["Score"].fillna(df["Score"].mean(), inplace=True)
df
```

```python
df2 = pd.DataFrame(data)
df2

df2["Score"] = df2['Score'].interpolate(method="linear")
df2
df2["Age"] = df2["Age"].interpolate(method="linear")
df2
```

### EDA using Bivariate and Multivarient  Analysis

```python
import pandas as pd
import seaborn as sns 
```

```python
tips = sns.load_dataset('tips')
titanic = pd.read_csv('train_sample.csv')
flights = sns.load_dataset('flights')
iris = pd.read_csv("iris_dataset.csv")
```

## 1. Scatterplot ( Numerical - Numerical )

```python
# Bivarint analys sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
sns.scatterplot(x=tips['total_bill'] , y=tips['tip'] , hue=tips['sex'] , style=tips['smoker'], size= tips['size'])
```

## 2. Bar Plot ( Numerical - Categorical )

```python
sns.barplot(x=titanic['Pclass'] , y=titanic['Fare'] , hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical )

```python
sns.boxenplot(x=titanic['Sex'] , y=titanic['Age'] ,hue = titanic['Survived'] )
```

## 4. Distplot ( Numerical - Categrorical)

```python
sns.displot(data=titanic, x="Age", hue="Survived", kind="kde")
```

## 5. HeatMap ( Categorical - Categorical)

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'] , titanic['Survived']))
```

## 6. ClusterMap ( Categrorical - Categorical )

```python
sns.clustermap(pd.crosstab(titanic['SibSp'] , titanic['Survived']))
```

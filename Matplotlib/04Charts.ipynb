
 # Bar charts are best for showing and comparing data that falls into different groups or categories. 
Here are some easy points to remember:

- Use bar charts for data divided into categories (like types of fruits, countries, or products).
- They make it simple to compare the size or amount of each category.
- Bar charts are clearer than line or scatter plots when your data is not about trends over time.
- Each bar shows a value for a category, making differences easy to see.

In short:  
Bar charts = Best for comparing groups or categories!

```python
import matplotlib.pyplot as plt

product = ['A','B' ,'C' ,'D']
salse = [1000 , 2000 , 3000 , 4000]

plt.barh(product , salse , color='orange' , label = 'sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparision')
plt.legend()
plt.show()
```

## Histogram specly used
- Histograms are best used for visualizing the distribution of numerical data.
- They help identify the frequency of data points within specified ranges (bins).
- Useful for detecting patterns such as skewness, modality (number of peaks), and outliers.
- Commonly used in statistics, quality control, and data analysis to understand data spread.
- Ideal for large datasets where individual data points are less important than overall distribution.

```python
import matplotlib.pyplot as plt

scores = [45 , 67 , 78 ,45, 67, 78,34,56, 67,7,88,89, 898 ]

plt.hist(scores , bins =5 , color = 'pink' , edgecolor ='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Student')
plt.title('Score distribution of Students')
plt.show()
```

### Basic view of matplotlip


```python
import matplotlib.pyplot as plt

x = ['mon' , 'tues' ,'wed' ,'thur' , 'fri'] # x-axis
y = [10 ,5 ,3 ,20 ,55]

plt.plot(x,y) # core plot commond hai 

plt.title('Bekery sales this week')

plt.xlabel('Day of the week')
plt.ylabel('sales per Day')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Bekery sales this week')
ax.set_xlabel('Day of the week')
ax.set_ylabel('sales per Day')
plt.show()
```

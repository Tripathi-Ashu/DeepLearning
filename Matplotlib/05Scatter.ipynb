```python
import matplotlib.pyplot as plt
hours_studies = [1 ,2 ,3,4, 5,6,7]
exam_scores = [50 ,55 , 60 , 70 ,75 ,80 , 85]

plt.scatter(hours_studies ,exam_scores , color='green' , marker = '^' , label = 'student Date')

plt.xlabel("Hours Studie")
plt.ylabel('Exam Score')
plt.title('Rekationship Between Study Time and  Exam Score ')
plt.legend()
plt.grid(True)
```

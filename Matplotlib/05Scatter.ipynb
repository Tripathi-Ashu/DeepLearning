```python
import matplotlib.pyplot as plt
hours_studies = [1 ,2 ,3,4, 5,6,7]
exam_scores = [50 ,55 , 60 , 70 ,75 ,80 , 85]

plt.scatter(hours_studies ,exam_scores , color='green' , marker = '^' , label = 'student Date')

plt.xlabel("Hours Studie")
plt.ylabel('Exam Score')
plt.title('Rekationship Between Study Time and  Exam Score ')
plt.legend()
plt.grid(True)
```

# 👉 Shortcut 

Category compare → Bar chart

Proportion → Pie chart

Trend over time → Line chart

Distribution → Histogram

Relationship between two variables → Scatter Plot ✅

```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3] ,[50 ,60,70] , color = 'blue' , label = 'Class A')
plt.scatter([1,2,3] ,[45,67,89] , color= 'orange' , label ='Class B')


plt.xlabel("Hours Studie")
plt.ylabel('Exam Score')
plt.title('Comparision of two classes')
plt.legend()
plt.grid(True)
```

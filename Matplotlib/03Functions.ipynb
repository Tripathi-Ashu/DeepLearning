```python
import matplotlib.pyplot as plt 

month = [1 , 2 ,3,4]
sales = [1000 , 2000 , 3000 , 8000]

plt.plot(month , sales , color='blue' , linewidth =2, linestyle = '--' ,marker ='o', label = '2025 salse data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title('Monthly sales data report')
plt.legend()
plt.grid(color = 'gray' , linestyle=':' , linewidth =1)
plt.xlim(1 ,4)
plt.ylim(0, 9000) 
plt.xticks([ 1,2,3,4] , ['M1' ,'M2' ,'M3' ,'M4'])

```

# ðŸ“Š Essential Subplot Functions in Matplotlib

Subplots in Matplotlib are used to create multiple plots within a single figure.  
Here are the **7 most important subplot functions** with explanations and examples:

---

## 1. ðŸ”— Shared X-Axis / Y-Axis
```python
fig, ax = plt.subplots(2, 1, sharex=True, sharey=True)
ax[0].plot([1,2,3], [10,20,30])
ax[1].plot([1,2,3], [15,25,35])
plt.show()

## 2. 
plt.subplot(1, 2, 1)
plt.plot([1,2,3], [1,2,3])
plt.title("Left")

plt.subplot(1, 2, 2)
plt.plot([1,2,3], [3,2,1])
plt.title("Right")

plt.show()

## 3. 
fig, ax = plt.subplots(2, 2)
fig.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()

## 4.
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [10,20,25,30])
ax.text(2, 25, "Important Point")
ax.annotate("Peak", xy=(3,25), xytext=(2,28),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

## 5.
fig, ax = plt.subplots(1, 2)
ax[0].plot([1,2,3],[1,4,9])
ax[0].set_title("Square")

ax[1].plot([1,2,3],[1,8,27])
ax[1].set_title("Cube")

fig.suptitle("Math Functions")
plt.show()

## 6.
fig, ax = plt.subplots(2, 2)
ax[0,0].plot([1,2,3],[1,2,3])
ax[0,1].plot([1,2,3],[3,2,1])
plt.show()

## 7.
fig, ax = plt.subplots(2, 2)
plt.tight_layout()
plt.show()



````python
#plt.subplot(nrows , ncols , index) 

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10 ,20 ,15, 25] 

plt.subplot(1,2,1) #1 row , 2 coloum , 1st subplot
plt.plot(x ,y)
plt.title('Line Chart')

plt.subplot(1 ,2 ,2) # 1row , 2 column , 2nd suplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()

````

````python
fig , ax = plt.subplots(1 ,2 , figsize =(10 ,5))

x = [1 ,2,3,4]
y = [10 ,15 , 20 ,8]

ax[0].plot(x ,y)
ax[0].set_title('Line Plot')

ax[1].bar(x ,y)
ax[1].set_title('Bar Chart')

fig.subtitle('Comparison of Line chart')
````
